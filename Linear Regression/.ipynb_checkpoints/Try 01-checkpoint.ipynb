```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('BreastCancer.csv')
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
data.dropna(inplace=True)
```

```python
#linear regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lean = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.columns
```

```python
X = data[['Age', 'Birth Year', 'Weight', 'Height', 'BMI', 'BreastFeeding',
       'Marital Status', 'Alcohol', 'Smoking', 'BreastCancerHistory',
       'Age at firstPeriod', 'MenstrualCycle']]
y = data['Cancer Positive']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X
```

```python
lean.fit(X_train,y_train)
```

```python
predictions = lean.predict(X_test)
```

```python
predictions
```

```python
y_test
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions))
```

```python
#Logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
log_predictions = log.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,log_predictions))
```

```python
#SVM
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
```

```python
svm_model.fit(X_train,y_train)
```

```python
svm_predictions = svm_model.predict(X_test)
```

```python
print(classification_report(y_test,svm_predictions))
```

```python
print(confusion_matrix(y_test,svm_predictions))
```

```python
#grid search to find the best parameters
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(),param_grid,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test,grid_predictions))
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
#random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_prediction = rfc.predict(X_test)
```

```python
print(classification_report(y_test,rfc_prediction))
```

```python
print(confusion_matrix(y_test,rfc_prediction))
```
