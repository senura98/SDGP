```python
#linear regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lean = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.columns
```

```python
X = data[['Age', 'BMI', 'BreastFeeding', 'Marital Status', 'Alcohol', 'Smoking',
       'BreastCancerHistory', 'Age at firstPeriod', 'MenstrualCycle']]
y = data['Cancer Positive']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lean.fit(X_train,y_train)
```

```python
predictions = lean.predict(X_test)
```

```python
predictions
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot(y_test-predictions)
```

```python
#logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
log_predictions = log.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,log_predictions))
```

```python
print(confusion_matrix(y_test,log_predictions))
```

```python
#SVM
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
```

```python
svm_model.fit(X_train,y_train)
```

```python
svm_predictions = svm_model.predict(X_test)
```

```python
print(classification_report(y_test,svm_predictions))
```

```python
print(confusion_matrix(y_test,svm_predictions))
```

```python
#gridsearch
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(),param_grid,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test,grid_predictions))
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
#Means Clustering 
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, 
                           centers=4, cluster_std=1.8,random_state=101)
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
km = KMeans(n_clusters=4)
```

```python
km.fit(X_train,y_train)
```

```python
km_predictions = km.predict(X_test)
```

```python
print(classification_report(y_test,km_predictions))
```

```python
print(confusion_matrix(y_test,km_predictions))
```
